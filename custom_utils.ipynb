```python
import os
from multiprocessing import Pool
from pydub import AudioSegment
from glob import glob
import io
import pandas as pd

```

```python
def get_files_by_type(extension):
    PATH = os.getcwd()
    files = [y for x in os.walk(PATH) for y in glob(os.path.join(x[0], f"*.{extension}"))]
    return files

```

```python
def get_duration(file_name):
    audio = AudioSegment.from_wav(file_name)
    return len(audio) / 1000.0  # Duration in seconds
```

```python
wav_files = get_files_by_type('wav')
txt_files = get_files_by_type('trans.txt')
```

```python
lines = []
ts_map = {}
for file in txt_files: 
    with open(file, 'r') as file:
        lines += file.readlines()
for line in lines:
    data = line.split(maxsplit=1) 
    ts_map[data[0]] = data[1]
```

```python
result = []
for file in wav_files:
    name = file.split('/')[-1].split('.wav')[0]
    result = result + [{
        "key": file, 
        "duration": get_duration(file),
        "text": ts_map[name].lower().replace('\n','')
    }]
```
